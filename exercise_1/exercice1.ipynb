### 2)

On utilise multivariate_normal pour générer des samples et représenter la distribution

```python
import numpy as np
import matplotlib.pyplot as plt

mean = [170, 70]
covariance_matrix = [[25, 10], [10, 36]]

n = 100
samples = np.random.multivariate_normal(mean, covariance_matrix, n)

plt.scatter(samples[:, 0], samples[:, 1])
plt.xlabel('Taille')
plt.ylabel('Poids')
plt.title('Bivariate Normal Distribution')
plt.show()
```

### 3)

Maintenant, pour des valeurs croissantes de n nouvs pouvons calculer la moyenne empirique des n premiers échantillons et représenter ainsi graphiquement la distance euclidienne entre la moyenne empirique et la valeur attendue.

```python
a = np.cumsum(samples, axis=0)
b = np.arange(1, n * 2 + 1).reshape(-1, 2)
empirical_averages = a / b
```

```python
expected_value = np.array(mean)

euclidean_distances = np.linalg.norm(empirical_averages - expected_value, axis=1)

plt.plot(np.arange(1, n + 1), euclidean_distances)
plt.xlabel('Nombre de samples (n)')
plt.ylabel('Distance Euclidienne')
plt.title('Convergence vers la valeur')
plt.show()
```

Les valeurs convergent bien.
